```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Data_Train.csv")
```

```python
dataset.head()
```

```python
dataset = dataset.drop("Date_of_Journey", axis = "columns")
dataset = dataset.drop("Source", axis = "columns")
dataset = dataset.drop("Destination", axis = "columns")
dataset = dataset.drop("Arrival_Time", axis = "columns")
dataset = dataset.drop(9039, axis = "rows")
dataset.head()
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
for i in range(len(x)):
  m = (int(x[i][2][0])*10 + int(x[i][2][1]))*60
  m += (int(x[i][2][3])*10 + int(x[i][2][4]))
  x[i][2] = m
print(x[0, :])
```

```python
for i in range(len(x)):
  m = 0
  f = 0
  for j in range(len(x[i][3])):
    if x[i][3][j] == "h":
      s = x[i][3][:j]
      m += int(s)*60
      f = j
    elif x[i][3][j] == "m":
      s = x[i][3][f+2:j]
      if s == "":
        m += 0
      else:
        m += int(s)
  x[i][3] = m
print(x[0, :])
```

```python
for i in range(len(x)):
  s = ""
  for j in range(len(x[i][1])):
    if x[i][1][j] != " " and x[i][1][j] != "?":
      s += x[i][1][j]
  x[i][1] = s
print(x[0, :])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, -1] = le.fit_transform(x[:, -1])
print(x[0, :])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(sparse_output=False), [0, 1, 4])], remainder="passthrough")
x = np.array(ct.fit_transform(x))
print(x[0, :])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x[:, -3:] = sc.fit_transform(x[:, -3:])
print(x[0, :])
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x, y)
```

```python
y_hat = regressor.predict(x)
np.set_printoptions(precision=2)
print(np.concatenate((y_hat.reshape(len(y_hat),1), y.reshape(len(y),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y, y_hat)
```
